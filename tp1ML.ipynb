```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

from sklearn import datasets
iriss =datasets.load_iris()

target = iriss.target
data =iriss.data #Pour un acces plus facile

fig = plt.figure(figsize=(10,5))
fig.subplots_adjust(hspace=0.5, wspace=0.5)
ax1 = plt.subplot(1,2,1)
clist = ['violet','yellow','blue']
colors = [clist[c]for c in target]

print(ax1.scatter(data[:,0],data[:,1],c=colors))
print(plt.xlabel('Longueur du sepal (cm)'))
print(plt.ylabel('largeur du seplal (cm)'))

ax2 = plt.subplot(1,2,2)
print(ax2.scatter(data[:,2],data[:,3],c=colors))
print(plt.xlabel('Longueur du petal (cm)'))
print(plt.ylabel('largeur du petal (cm)'))

for ind , s in enumerate(iriss.target_names):
    plt.scatter([],[],label=s, color=clist[ind])
    plt.legend(scatterpoints=1,frameon=True,labelspacing=1,bbox_to_anchor=(1.8,0.5),loc="center right" ,title='Espces')
plt.plot()
```
