```python
from sklearn import datasets
import matplotlib.pyplot as plt ##import de la bib de affichage
##prepare les donner
iris=datasets.load_iris()
data=iris.data
##les donner pour dessiner les axe 0 et 1  pour les sepale et 2 , 3 pitale
x=data[:,2]
y=data[:,3]


axe=plt.axes()##dessiner le 2 axes
axe.grid()##dessiner en forme de grid
plt.scatter(x,y)## les donnes seront au forme de nuage des pts


##dessiner le ligne de regression
from scipy import stats

slope,intercept,r,p,std_err=stats.linregress(x,y)##applique la fonction lineregress(x,y) pour dessiner la ligne (elle retourne 5 val ) #a=slope , b=intercept

##en prepare la f=ax+b
def predict(x):
    return slope*x+intercept
##preparer la ligne on calculant f(x)
fitline=predict(x)# cette fonction essayer de estime le y
## dessiner la ligne en rouge on donnant X pour calculer f(x)=fitline
plt.plot(x,fitline,c='r')

```
